```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting
%matplotlib inline
```

```python
# Load the dataset (Replace with the correct CSV file if downloaded)
url = 'https://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
```

```python
# Example: Simulate sample population data for visualization
data = {
    'Country': ['India', 'China', 'USA', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1417173173, 1425671352, 339996563, 277534122, 240485658,
                   216422446, 223804632, 169828911, 144444359, 127504125]
}
df = pd.DataFrame(data)
df
```

```python
# Bar chart to show population distribution by country
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Population', data=df, palette='viridis')
plt.title('Population Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
